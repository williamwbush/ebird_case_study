```python
df_species_count = df.groupby("COMMON NAME", as_index=False).count()
df_species_count.head(5)
```

```python
df["OBSERVATION COUNT INT"] = df["OBSERVATION COUNT"]

df["OBSERVATION COUNT INT"].replace(to_replace="X", value=0, inplace=True)

df["OBSERVATION COUNT INT"] = df["OBSERVATION COUNT INT"].astype(int)
```

```python
df_obs_count = df.groupby("COMMON NAME", as_index=False).sum()[["COMMON NAME", "OBSERVATION COUNT INT"]].sort_values("OBSERVATION COUNT INT", ascending=False).reset_index(drop=True)
df_obs_count
```

```python


# trend of observations per day in 2020

daily_obs = df.groupby("OBSERVATION DATE", as_index=False).sum()[["OBSERVATION DATE", "OBSERVATION COUNT INT"]]
daily_obs_18to20 = daily_obs[(daily_obs["OBSERVATION DATE"] > "2017-12-31") & (daily_obs["OBSERVATION DATE"] < "2021-01-01")]
daily_obs_18to20

daily_obs_20 =  daily_obs[(daily_obs["OBSERVATION DATE"] > "2019-12-31") & (daily_obs["OBSERVATION DATE"] < "2021-01-01")]
```

```python
obs_date = list(daily_obs_18to20["OBSERVATION DATE"])
obs_count = list(daily_obs_18to20["OBSERVATION COUNT INT"])

plt.plot(obs_date, obs_count)
plt.show()
```

```python
obs_date_20 = list(daily_obs_20["OBSERVATION DATE"])
obs_count_20 = list(daily_obs_20["OBSERVATION COUNT INT"])

plt.plot(obs_date_20, obs_count_20)
plt.show()
```
