```python
df_all_reported = df[df['ALL SPECIES REPORTED'] == 1]
df_checklist = df_all_reported[['OBSERVATION DAY OF YEAR', 'SAMPLING EVENT IDENTIFIER', 'COMMON NAME']]
df_checklist_count = df_checklist.groupby('OBSERVATION DAY OF YEAR', as_index=False).nunique().sort_values('OBSERVATION DAY OF YEAR').reset_index(drop=True)
```

```python
chimney_swift_df = df_checklist[df_checklist['COMMON NAME'] == 'Chimney Swift']
chimney_swift_checklist_count = chimney_swift_df.groupby('OBSERVATION DAY OF YEAR', as_index=False).count().sort_values('OBSERVATION DAY OF YEAR').reset_index(drop=True)
chimney_swift_checklist_count['CHIMNEY SWIFT CHECKLIST SUM'] = chimney_swift_checklist_count['COMMON NAME']
```

```python
total_and_cs_checklist_counts = df_checklist_count.merge(chimney_swift_checklist_count, on = "OBSERVATION DAY OF YEAR", how = "outer")
total_and_cs_checklist_counts["CHIMNEY SWIFT CHECKLIST SUM"] = total_and_cs_checklist_counts["CHIMNEY SWIFT CHECKLIST SUM"].fillna(0)
total_and_cs_checklist_counts
```

```python
total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY'] = total_and_cs_checklist_counts['CHIMNEY SWIFT CHECKLIST SUM'] / total_and_cs_checklist_counts['SAMPLING EVENT IDENTIFIER_x']
total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY'] = round(total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY'] * 100)
```

```python
series = total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY']
rolling_ave_5 = [0,0]
for i in range(2,len(series)-2):
    rolling_ave_5.append(sum(series[i-2:i+3]) / 5)
rolling_ave_5 = rolling_ave_5 + [0,0]
total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY 5 DAY RA'] = rolling_ave_5
```

```python
rolling_ave_7 = [0,0,0]
for i in range(3,len(series)-3):
    rolling_ave_7.append(sum(series[i-3:i+4]) / 7)
rolling_ave_7 = rolling_ave_7 + [0,0,0]
total_and_cs_checklist_counts['CHIMNEY SWIFT FREQUENCY 7 DAY RA'] = rolling_ave_7
```

```python
total_and_cs_checklist_counts_daily_7RA = total_and_cs_checklist_counts[['OBSERVATION DAY OF YEAR','CHIMNEY SWIFT FREQUENCY','CHIMNEY SWIFT FREQUENCY 7 DAY RA']]
total_and_cs_checklist_counts_daily_7RA
total_and_cs_checklist_counts_daily_7RA.to_csv('total_and_cs_checklist_counts_daily_7RA.csv') 
```

```python
# demonstrate irruptive behavior of common redpoll in Illinois
df_all_reported = df[df['ALL SPECIES REPORTED'] == 1]
df_checklist_date = df_all_reported[['OBSERVATION DATE','OBSERVATION DAY OF YEAR', 'SAMPLING EVENT IDENTIFIER', 'COMMON NAME']]
df_checklist_date_count = df_checklist_date.groupby(['OBSERVATION DATE','OBSERVATION DAY OF YEAR'], as_index=False).nunique().sort_values('OBSERVATION DATE').reset_index(drop=True)
df_checklist_date_count = df_checklist_date_count[(df_checklist_date_count['OBSERVATION DATE'] > "1999-12-31") & (df_checklist_date_count['OBSERVATION DATE'] < "2021-02-01")]
```

```python
common_redpoll_df = df_checklist_date[df_checklist_date['COMMON NAME'] == 'Common Redpoll']
common_redpoll_checklist_count = common_redpoll_df.groupby(['OBSERVATION DATE','OBSERVATION DAY OF YEAR'], as_index=False).count().sort_values('OBSERVATION DATE').reset_index(drop=True)
common_redpoll_checklist_count['COMMON REDPOLL CHECKLIST SUM'] = common_redpoll_checklist_count['COMMON NAME']

common_redpoll_checklist_count = common_redpoll_checklist_count[(common_redpoll_checklist_count['OBSERVATION DATE'] > "1999-12-31") & (common_redpoll_checklist_count['OBSERVATION DATE'] < "2021-02-01")]
common_redpoll_checklist_count
```

```python
total_and_cred_checklist_counts = df_checklist_date_count.merge(common_redpoll_checklist_count, on = "OBSERVATION DATE", how = "outer")
total_and_cred_checklist_counts["COMMON REDPOLL CHECKLIST SUM"] = total_and_cred_checklist_counts["COMMON REDPOLL CHECKLIST SUM"].fillna(0)
total_and_cred_checklist_counts
```

```python
total_and_cred_checklist_counts['COMMON REDPOLL FREQUENCY'] = total_and_cred_checklist_counts['COMMON REDPOLL CHECKLIST SUM'] / total_and_cred_checklist_counts['SAMPLING EVENT IDENTIFIER_x']
total_and_cred_checklist_counts['COMMON REDPOLL FREQUENCY'] = round(total_and_cred_checklist_counts['COMMON REDPOLL FREQUENCY'] * 100, 3)
```

```python
# splitting up data by year

years = np.arange(2001,2022)
cred_yearly = []

for year in years:
    cred_yearly.append(total_and_cred_checklist_counts[(total_and_cred_checklist_counts['OBSERVATION DATE'] > f"{year-1}-06-30") & (total_and_cred_checklist_counts['OBSERVATION DATE'] < f"{year}-07-01")].reset_index(drop=True))  

cred_yearly

for season in cred_yearly:
    cred_rolling_ave_7 = [0,0,0]
    for i in range(3,len(season['OBSERVATION DATE'])-3):
        cred_rolling_ave_7.append(sum(season['COMMON REDPOLL FREQUENCY'][i-3:i+4]) / 7)
    cred_rolling_ave_7 = cred_rolling_ave_7 + [0,0,0]
    season['FREQUENCY 7 DAY RA'] = cred_rolling_ave_7

cred_yearly
```

```python
c = 2001
for year in cred_yearly:
    year.to_csv(f'cred_frequency_{c}.csv')
    c += 1
```
